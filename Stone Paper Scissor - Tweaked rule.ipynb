```python
def game3():
    
    counta=0
    countb=0

    while counta<5 and countb<5:

        inpa=input('Player A - Enter stone,paper or scissor :')
        inpa=inpa.lower().replace(' ','')
        inpb=input('Player B - Enter stone,paper or scissor :')
        inpb=inpb.lower().replace(' ','')
        
        if inpa==inpb:
            counta=counta+1
            countb=countb+1
            print('It is a DRAW')

        if inpa=='stone' and inpb=='paper':
            countb=countb+1
            print('Point to B.Score of A ',counta,', Score of B',countb)

        if inpa=='paper' and inpb=='scissor':
            countb=countb+1
            print('Point to B.Score of A ',counta,', Score of B',countb)

        if inpa=='scissor' and inpb=='stone':
            countb=countb+1
            print('Point to B.Score of A ',counta,', Score of B',countb)

        if inpb=='scissor' and inpa=='stone':
            counta=counta+1
            print('Point to A.Score of A ',counta,', Score of B',countb)

        if inpb=='stone' and inpa=='paper':
            counta=counta+1
            print('Point to A.Score of A ',counta,', Score of B',countb)

        if inpb=='paper' and inpa=='scissor':
            counta=counta+1
            print('Point to A.Score of A ',counta,', Score of B',countb)


    print('Total count of Player A is',counta)
    print('Total count of Player B is',countb)

    if counta==5 and countb==5:
        print('Both win')
    elif counta==5:
        print('A wins the game')
    else:
        print('B wins the game')
        
        
game3()
```
